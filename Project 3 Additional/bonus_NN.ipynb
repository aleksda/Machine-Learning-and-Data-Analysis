```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
```

```python
from sklearn.neural_network import MLPRegressor

def run_MLP_boot(nr_of_hidden_nodes = [[10], [10, 10]], boot_model = None, lr = 0.01, alpha = 0.1):
    
    mse_list, bias_list, var_list = {}, {}, {}
    xaxis = [len(i) for i in nr_of_hidden_nodes]

    for i in nr_of_hidden_nodes:
        
        mlp_model = MLPRegressor(hidden_layer_sizes = i, alpha = alpha, learning_rate_init = lr)
        
        t_tilde, t_pred = boot_model.simulate(mlp_model)
        mse, bias, var  = boot_model.decompose(t_pred)

        mse_list[i]  = mse
        bias_list[i] = bias
        var_list[i]  = var 

    plt.style.use("seaborn-darkgrid")
    plt.title('Bias-variance tradeoff for MLP')
    plt.xlabel('Number of hidden layers'); plt.ylabel('Error/Bias/Variance')
    plt.plot(xaxis, mse_list.values(),  '--o', c='blue', label='Mse')
    plt.plot(xaxis, bias_list.values(), '--o', c='orange', label='Bias')
    plt.plot(xaxis, var_list.values(),  '--o', c='green', label='Variance')
    plt.legend(loc='best')
    plt.savefig('NN_HN.png')
    
    return mse_list, bias_list, var_list
```

```python
import frankefunction as fr
from bootstrap import *

z = fr.Franke(0, 0.2)
z_noisy, xx, yy = z.franke_function(random = False, precision = 0.05, noise = False) 
print(xx.shape, yy.shape)

boot = Bootstrap(xx, yy, z_noisy, seed = None, scale_target=True)

hl = [(10,), (10,)*3, (10,)*5, (10,)*7, (10,)*9, (10,)*11]
#hn = [(5,), (10,), (30,), (50,), (100,), (200,), (500,)]

mse_linreg, bias_linreg, var_linreg = run_MLP_boot(nr_of_hidden_nodes = hl, boot_model = boot, alpha = 0.1)
#mse_linreg, bias_linreg, var_linreg = run_MLP_boot(nr_of_hidden_nodes = hn, boot_model = boot, alpha = 0.1)
```
